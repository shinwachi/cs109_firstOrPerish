## Introduction

### Background

"Publish or Perish" is a common phrase in academia to emphasize the importance of frequent journal publication to improve the individual scientist's career prospect (http://en.wikipedia.org/wiki/Publish_or_perish).  A scientist is often judged by her peer-reviewed article as record of her performance in research.  

However, not all authorships are created equal. "First author" (or "lead author" http://en.wikipedia.org/wiki/Lead_author) refers to the main contributor to the research that went into the publication of a research article, and gets the most credit.  Another convention in biomedical journal article is the concept of "last author", which indicates the most senior member of the lab which has produced the paper, and it is often a faculty of the institution.  It is safe to say that at least in academia, a position as a senior member of the lab is considered to be success in a scientist's career.  In addition, publication in a prestigious academic journals is considered to be more valuable, and indicates the potential influence of the article published in it.  

### Motivation

There are previous studies which uses bibliographic records to analyze the behaviors of scientists.  For example, co-authorship network is used to study the relationship of collaborations between scientists (http://www.sciencedirect.com/science/article/pii/S1751157711000630).  To our knowledge, the importance of author order has not been fully explored.  Our research will take this into consideration, with the goal of evaluating the predictive values of bibliographic records with author order to determine success in scientific careers.

Our goal is to obtain individual author's full publication records from Medline database as a representation of their entire academic career.  We use the author order to determine the scientists' annual productivity (first- and middle-authorships), as well as success in becoming a senior researcher/faculty of the lab (last author publications).  Finally, we will determine if these bibliographic data with author order information (as well as the impact factor of the journals) is able to predict the author success, and to explore how faculty salary may be influenced by past research productivity as measured by the publication record.



## Results

We have obtained and analyzed 1.5 million authors bibliographic record with individual roles extrapolated with the author positions.  The bibliographic record was combined with Eigenfactor (journal impact factor substitute) and faculty salary data.  Using exploratory data analysis (EDA), we were surprised to find that there was no strong relationship between author frequency nor the quality of the journals the the author published the articles to whether or not the author became the last author, nor have any bearing to the salaries they received.  We observed a very weak relationship between first author frequency to becoming the last author or not, with last authors tended to have more frequent first author papers in the earlier part of their career.

#### Author Filtering

The way authors are identified in Medline, which is by forename and surname, does not guarantee
unique identification.  

--------

** Example 1: **

Author A and B have the same name, and their career overlaps.

`
 1984                      2001
  AAAAAAAAAAAAAAA
      BBBBBBBBBBBBBBBBBBBBBBB
      OOOOOOOOOOO              <--- overlap
  YYYYYYYYYYYYYYYYYYYYYYYYYYY  <--- artifact
`

For the duration of the overlapping period ('O'), author Y (which is actually two people) will look very productive by increased frequency of authorship (FRQFA or FRQTA).

------

** Example 2: **

Author C and D have the same name, and their career years ar far apart.

`
1952                                      2011
CCCCCCCCCC
                                     DDDDDDDD
XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX <--- artifact
`

While both C and D did not have a long career, author X (which is acutally two people) will look like they have had a very long career (and low productivity).

---------

Therefore, it was important to look at the distribution of calculated author factors and to filter out odd entries due to either multiple authors having the same name, either by pure coincidence or by error.  For example, we have excluded excessive career length (> 70 years), considering the likely age of first publication (in the 20s at the earliest) and human life span. Another is to limit the frequency of first-author paper (< 25), since it is very unlikely that a researcher will be able to publish more than around two papers a month as a first author in a year.

We had required the author having career length over 3 years to reduce the authors with a very brief career, because anything shorter will not give meaningful estimate of publication frequency.

We also excluded authors whose first paper was before 1983, this year was chosen because this year heralded the era of modern biochemistry with the invention of PCR (polymerase chain reaction).

## NML baseline XML Parsing and Data Extraction
2014 Medline Baseline dataset (medline14) was downloaded from National Library of Medicine (ftp.nlm.nih.gov, using leasing license for access, on November, 2014.  http://www.nlm.nih.gov/databases/journal.html).  The collection consists of a set of 746 xml files, with .gz compressed size of approximately 13 GB.


```python
# preamble
import os, csv, gzip, json, re
import codecs
import unicodedata
from bs4 import BeautifulSoup
from bs4 import SoupStrainer
from collections import OrderedDict
```

The following are the support functions:

```python
def getFields(leafxmlfragment, fieldlist):
    answerdict = {}
    for field in fieldlist:
        # find all text with field tags
        x = bsFragment(leafxmlfragment).find_all(field)
        if x: # if field text exists, get rid of unicode
            answerdict[field] = unicodedata.normalize('NFKD',unicode(x[0].string)).encode('ascii','ignore') 
        else: # blank default value
            answerdict[field] = ''
    return answerdict

def bsFragment(xmlfragment):
    ''' ensures that a fragment is a beautiful soup (bs) fragment object '''
    return BeautifulSoup(str(xmlfragment))

def bs_findall(xmlfragment, *arg):
    ''' Recursive tag search (arg is variable arity) '''
    xmlfragment = bsFragment(xmlfragment)
    # go through nested tags to get to the last tag, collect the last tag
    answer = []
    if arg != ((),): # if not an empty arg
        searchterm = arg[0]
        for x in xmlfragment.find_all(searchterm):
            if len(arg) > 1:
                # recursion into next arg (arg[1:])
                answer.extend(bs_findall(str(x),arg[1:] ))
            else:
                answer.append(str(x))
    else:
        return [] 
    return answer

def getDate(xmlfragment):
    ''' get article date, at least the years '''
    fields = 'year month day'
    x2     = getFields(bs_findall(xmlfragment, 'articledate'),fields.split())
    medlinedate_yearregex = re.compile(r'\d+')
    if not x2['year']:
        # Then look for another date fields
        x = bs_findall(xmlfragment, 'MedlineDate')
        try:
            yr = medlinedate_yearregex.findall(getFields(x, ['medlinedate'])['medlinedate'])[0]
        except:
            # give up finding years
            yr = ''
        x2 = {'year':yr, 'month':'', 'day':''}
        
    if not x2['year']:
        x = bs_findall(xmlfragment, 'pubdate')
        fields = 'year month day'.split()
        x2 = getFields(x,fields)
    
    return getFields(x,fields)
    

def getAuthors(xmlfragment):
    ''' extract author information '''
    authordicts = []
    for authorxml in bs_findall(xmlfragment, 'authorlist','author'):
        fields     = 'lastname forename initials affiliation'
        authordict = getFields(authorxml, fields.split())
        authordicts.append(authordict)
    return authordicts


def getJournalInfo(xmlfragment):
    ''' get journal info '''
    fields = 'country medlineta nlmuniqueid issnlinking'.split()
    try:
        x = bs_findall(xmlfragment, 'medlinejournalinfo')[0]        
    except:
        x = ""
    return getFields(x, fields)


def processArticle(xmlfragment):
    ''' 
    consolidate all the functions above and 
    executes them on the xml fragment to retrieve 
      date, 
      journal, and
      authors
    '''
    #               key      functions
    fncDict = dict(date    = getDate, 
                   journal = getJournalInfo, 
                   authors = getAuthors)
    return dict([(k, vfnc(xmlfragment)) for k,vfnc in fncDict.items()])
    
```


The following code takes a medline xml file and extracts the information. 

** Warning: The real dataset took hours to process. This example file will take a few minutes as well. **

```python
inpath  = 'data/medline14n0746.xml.gz'
outpath = 'data/medline14n0746.txt.gz'

originalfilename = os.path.split(inpath)[1]

f = gzip.open(inpath, 'rb')
file_content = f.read()
f.close()

# split citations to chunk xml parsing (rely on closing tags as it has no other properties)
k = file_content.split('</MedlineCitation>')

f = gzip.open(outpath, 'w')
for entryNumber, xmlfragment in enumerate(k):
    if entryNumber%500 == 0:
        print "%done:", float(entryNumber)/float(len(k)) * 100.0, 'done', entryNumber, "total",  len(k)
    entryId = "%s:%d"%(originalfilename, entryNumber)
    try:
        data = processArticle(xmlfragment)
    except:
        data = {}
    data['entryid'] = entryId
    f.write("%s\n"%str(json.dumps(data)))

f.close()
```

### Consolidation of Author Information

The resultant JSON file had a format which included one article per row.  Because we wanted to get information about individual authors, we have transformed the format to one author per row, with multiple articles.


##### File-wise consolidation



e.g.,

`
File_1:
article_1: [auth_A, auth_B]   <<<
article_2: [auth_B, auth_C]   <<<
`

is then transformed to

`
File_1:
auth_A: [article_1]
auth_B: [article_1, article_2] <<<
auth_C: [article_2]
`

```python
import gzip, json, csv
###################################################################
def jsonLoads(jsonString):
    '''
    # helper function to convert unicode to str
    # used to decode SOAP reply
    # http://stackoverflow.com/questions/956867/how-to-get-string-objects-instead-of-unicode-ones-from-json-in-python

    replace:
        json.loads(result.PE_executeResponse.Response, object_hook=decode_dict)
    with:
        jsonLoads(result.PE_executeResponse.Response)
    '''
    def _decode_list(data):
        rv = []
        for item in data:
            if isinstance(item, unicode):
                item = item.encode('utf-8')
            elif isinstance(item, list):
                item = _decode_list(item)
            elif isinstance(item, dict):
                item = _decode_dict(item)
            rv.append(item)
        return rv

    def _decode_dict(data):
        rv = {}
        for key, value in data.iteritems():
            if isinstance(key, unicode):
                key = key.encode('utf-8')
            if isinstance(value, unicode):
                value = value.encode('utf-8')
            elif isinstance(value, list):
                value = _decode_list(value)
            elif isinstance(value, dict):
                value = _decode_dict(value)
            rv[key] = value
        return rv
    return json.loads(jsonString, object_hook=_decode_dict)
###################################################################

def getAuthorInfo(d2, authorData, count):
    # get author list ([] if it doesn't exist)
    if 'authors' in d2.keys():
        authors = d2['authors']
    else:
        authors = []

    # filter authors with no last name
    authors   = [author for author in authors if author['lastname']]
    authCount = len(authors)

    processedAuthor = []
    for authPos, author in enumerate(authors, start=1):
        
        # author unique ID
        authorKey = (".".join([author['forename'].replace(" ","_"), author['lastname']])).lower()

        # assign first author affiliation if nothing else exists
        if author['affiliation']:
            affiliation = author['affiliation']
        else:
            affiliation = authors[0]['affiliation']


        nlmuniqueid  = d2['journal']['nlmuniqueid']
        year         = d2['date'   ]['year'       ]
        
        article = dict(
            nlmuniqueid  = nlmuniqueid,
            affiliation  = affiliation ,
            authPos      = authPos,
            authCount    = authCount,
            year         = year
            )

        try:
            if not authorKey in processedAuthor: # check for duplicate author in article
                if authorKey not in authorData.keys():
                    authorData[authorKey] = []
                
                authorData[authorKey].append(article)
            else:
                pass # skip duplicate author

        except:
            print "Exception: ", article , " :Author: " , authorKey , " :Count: " , count
        
        processedAuthor.append(authorKey)
    return
```

```python
import gzip, csv, os, re, ast
import numpy as np

# pattern to parse medline___.txt files
#                     author,#field,
pattern = re.compile("([^,]*),[^\[]*(\[.*\])[^\]]*")

def tryfunc(xfunc, xs,  valOnException=None):
    '''
    Tries a function and returns valOnException on exception
    '''
    try:
        return xfunc(xs)
    except:
        return valOnException

def mean(x):
    if x:
        return np.average(x)
    else:
        return None


#This funciton calculates all summary Stats
def summaryDictionary(allJournals):
    '''
    FFA: first first-author year:  1879.0
    FLA: first last-author year:  1968.0
    LAA: last all-author year:  1998.0
    TFA: total first author paper:  28
    TLA: total last author paper:  6
    TAA: total paper (all author):  41
    TCL: total career legth =  LAA - FFA:  119.0
    FCL: faculty career length = LAA - FLA:  30.0
    SCL: non-faculty scientist career length = TLA-FCL:  1849.0
    TCL2: total career length = SCL + FCL:  1879.0
    FRQFA: frequency of first author = TFA/SCL:  0.0151433207139
    FRQTA: frequency of all authorship = TAA/TCL:  0.344537815126
    FAEF: average first-author eigenfactor
    MAEF: average middle-author eigenfactor
    LAEF: average last-author eigenfactor
    '''

    pubsummary = []
    for entry in allJournals:
        dictEn = dict(entry)
        year = dictEn['year']
        if year:
            year = float(year)        
        isFirstAuthor = dictEn['authPos'] == 1
        isLastAuthor  = dictEn['authPos'] == dictEn['authCount']
        nlmuniqueid   = dictEn['nlmuniqueid']
        affiliation   = dictEn['affiliation']
        try:
            ef = eigenfactordict[nlmuniqueid]
        except:
            ef = None
        
        pubsummary.append(dict(year=year,isFirstAuthor=isFirstAuthor, isLastAuthor=isLastAuthor, ef = ef, affiliation = affiliation))
        
    taap_total_paper              = len(pubsummary)
    tfap_total_first_author_paper = len([x for x in pubsummary if x['isFirstAuthor']])
    tlap_total_last_author_paper  = len([x for x in pubsummary if x['isLastAuthor']])
    
    # filter for publication with actual years
    pubsummary_years = [x for x in pubsummary if x['year']]
    
    laay_last_all_author_paper_year    = tryfunc(max, [x['year'] for x in pubsummary_years])
    ffay_first_first_author_paper_year = tryfunc(min, [x['year'] for x in pubsummary_years if x['isFirstAuthor']])
    flay_first_last_author_paper_year  = tryfunc(max, [x['year'] for x in pubsummary_years if x['isLastAuthor']])
    
    
    # average first  author EF 
    faef_first_author_ef  = mean([x['ef'] for x in pubsummary if     x['isFirstAuthor'] and x['ef']])
    # average middle author EF
    maef_middle_author_ef = mean([x['ef'] for x in pubsummary if not x['isFirstAuthor'] and not x['isLastAuthor'] and x['ef']])
    # average last   author EF
    laef_middle_author_ef = mean([x['ef'] for x in pubsummary if     x['isLastAuthor']  and x['ef']])
    # scientist (non-last author) EF
    saef_all_author_ef    = mean([x['ef'] for x in pubsummary if not x['isLastAuthor']  and x['ef']])
    
    # get list of affiliations for last-author papers
    zaffs = ','.join(['"""%s"""'%x['affiliation'] for x in pubsummary if x['affiliation'] and x['isLastAuthor']])
    
    
    # total caeer length (tcl = laay - ffay)
    if laay_last_all_author_paper_year and ffay_first_first_author_paper_year:
        tcl_total_career_length = laay_last_all_author_paper_year - ffay_first_first_author_paper_year
    else:
        tcl_total_career_length = 0

    # faculty caeer length (fcl = laay - flay)
    if laay_last_all_author_paper_year and flay_first_last_author_paper_year:
        fcl_faculty_career_length = laay_last_all_author_paper_year - flay_first_last_author_paper_year
    else:
        fcl_faculty_career_length = 0
    
    # scientist career length
    scl_scientist_career_length = tcl_total_career_length - fcl_faculty_career_length
    
    
    #Can't divide by 0, so force the career length minimum to 1 year, (which is reasonable)
    #change defaults here if they are not changed from above
    if scl_scientist_career_length == 0:
        scl_scientist_career_length = 1
    if tcl_total_career_length == 0:
        tcl_total_career_length = 1
    
    summary = dict(
            ffay   = ffay_first_first_author_paper_year,
            flay   = flay_first_last_author_paper_year,
            laay   = laay_last_all_author_paper_year,
            tfap   = tfap_total_first_author_paper,
            tlap   = tlap_total_last_author_paper,
            taap   = taap_total_paper,
            tcl   = tcl_total_career_length,
            fcl   = fcl_faculty_career_length,
            scl   = scl_scientist_career_length,
            frqfa = tfap_total_first_author_paper/scl_scientist_career_length,
            frqta = taap_total_paper/tcl_total_career_length,
            faef  = faef_first_author_ef,
            maef  = maef_middle_author_ef,
            laef  = laef_middle_author_ef,
            saef  = saef_all_author_ef,
#             zaffs = zaffs,
        )
    return summary
```

```python


header = "College	Full professors	Associate professors	Assistant professors	Instructors	Student-fac. ratio"
rules = [{'pattern':header, 
          'result':True}, 
         {'pattern':r"Showing \d+–\d+ of [\d\,]+ colleges", 
          'result':False}]

def applyMatchRules(state, rules, s):
    ''' pattern matching function '''
    for rule in rules:
        if re.search(rule['pattern'], s):
            return rule['result']
    return state


answer = []
state = False
for row in rawdata:
    case = (state, applyMatchRules(state,rules,row))
    if   (False, True) == case: 
        state = True
    elif (True, False) == case: 
        state = False
    elif state == True:
        answer.append(dict(zip(header.split('\t'), row.replace('\n','').split('\t'))))


sfratio = "Student-fac. ratio"
faculties = "Assistant professors	Associate professors	Full professors	Instructors".split('\t')


def toInt(x):
    if x == '':
        return None
    else:
        return int(x)

# hint to replace $ and ,
# http://stackoverflow.com/questions/17667606/efficiently-converting-strings-to-appropriate-numeric-types-in-pandas-dataframe
facsaldf = DataFrame(answer)
for faculty in faculties:
    
    facsaldf[faculty] = [toInt(x) for x in list(facsaldf[faculty].str.replace(r'[$,]', ''))] #facsaldf[faculty].str.replace(r'[$,]', '')

# facsaldf.to_dict()


# list(facsaldf[faculty].str.replace(r'[$,]', ''))
facsaldf.to_csv('data/faculty_salary_2012.csv')
facsaldf.head()
```
