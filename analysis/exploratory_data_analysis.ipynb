```python
# http://nbviewer.ipython.org/github/cs109/content/blob/master/labs/lab4/Lab4full.ipynb
from sklearn.linear_model import LogisticRegression
reg=1000.
clf4 = LogisticRegression(C=reg)
clf4.fit(d.logfrqfa.reshape(-1,1), d.la)

```

```python
tempsnew=np.linspace(0., 40., 1)
probs = clf4.predict_proba(tempsnew.reshape(-1,1))[:, 1]
predicts = clf4.predict(tempsnew.reshape(-1,1))
```

```python
plt.scatter(d.logfrqfa[minisample], d.la[minisample])
axes=plt.gca()
axes.grid(False)
# remove_border(axes)
plt.plot(tempsnew, probs, marker='s')
plt.scatter(tempsnew, predicts, marker='s', color="green")
```
