```python
#    FFA: first first-author year:  1879.0
#    FLA: first last-author year:  1968.0
#    LAA: last all-author year:  1998.0
#    TFA: total first author paper:  28
#    TLA: total last author paper:  6
#    TAA: total paper (all author):  41
#    TCL: total career legth =  LAA - FFA:  119.0
#    FCL: faculty career length = LAA - FLA:  30.0
#    SCL: non-faculty scientist career length = TLA-FCL:  1849.0
#    TCL2: total career length = SCL + FCL:  1879.0
#    FRQFA: frequency of first author = TFA/SCL:  0.0151433207139
#    FRQTA: frequency of all authorship = TAA/TCL:  0.344537815126
#    FAEF: average first-author eigenfactor
#    MAEF: average middle-author eigenfactor
#    LAEF: average last-author eigenfactor
#NOTE: I am using the following file: authorStat_small.txt.gz
#If you run all cells the last cell wil have the data after outliers
# This data accounts for the following data and thresholds
#* FCL - faculty career length = LAA - FLA:  30.0 
#    * Set a max threshold of 50 years 
#    * this eliminated 0.00172724221399 of the total data set
#
#* SCL - non-faculty scientist career lengt
#    * Set a max threshold of 70 years 
#    * this eliminated 0.00279090181428 of the total data set
# 
#* TCL - total career legth
#    * Set a max threshold of 70 years 
#    * this eliminated 0.00376176998685 of the total data set
#
#* TAAP - total paper (all author)
#    * Set a max threshold of 150 papers 
#    * this eliminated 0.011366536878 of the total data set
#
#*  FRQFA: frequency of first author
#    * Set a max threshold of 25
#    * this eliminated 0.000186133546414 of the total data set
# we started with 3,631,801 records
# After all these thresholds are met about 1.59% of data is excluded leaving us with 3,574,049 records 
# columns:
#     General: author
#     Decimals: saef, laef, maef, frqta, frqfa, faef
#     Year: ffay, flay, laay
#     Frequency: scl, tlap, tcl, taap, tfap, fcl
```

```python
import gzip, json, csv, os
import pandas as pd
import ast
from datetime import timedelta, date
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
rcParams['figure.figsize'] = (20, 6)

#infileDirPath = '../../../../../Dropbox/cs109_datasci_share_2/authout/'
infileDirPath = '../../../../../Desktop/'
#infileDirPath = '/vagrant/cs109_firstOrPerish/data_munging/'
allFiles = []
for file in os.listdir(infileDirPath):
    if file.endswith("small.txt.gz"):
        allFiles.append(file)
```

```python
print allFiles[:1]
for signFile in allFiles[:1]:
    signFile = infileDirPath + signFile
    print signFile
    outfile = gzip.open( signFile, 'r')
    reader = pd.read_csv(outfile)
```
