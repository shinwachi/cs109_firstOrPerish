```python
import gzip, json, csv, os
import pandas as pd
import ast
from datetime import timedelta, date
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
rcParams['figure.figsize'] = (20, 6)

#infileDirPath = '../../../../../Dropbox/cs109_datasci_share_2/authout/'
infileDirPath = '../../../../../Desktop/'
#infileDirPath = '/vagrant/cs109_firstOrPerish/data_munging/'
allFiles = []
for file in os.listdir(infileDirPath):
    if file.endswith("small.txt.gz"):
        allFiles.append(file)
```

```python
print allFiles[:1]
for signFile in allFiles[:1]:
    signFile = infileDirPath + signFile
    print signFile
    outfile = gzip.open( signFile, 'r')
    reader = pd.read_csv(outfile)
```

```python
# columns:
#     General: author
#     Decimals: saef, laef, maef, frqta, frqfa, faef
#     Year: ffay, flay, laay
#     Frequency: scl, tlap, tcl, taap, tfap, fcl
reader.tail()
```

```python
decimalColumns = ['saef', 'laef', 'maef', 'frqta', 'frqfa', 'faef']
for columns in decimalColumns:
    subData = reader[columns]
    subData = pd.DataFrame(subData)
    allPoints = []
    for item in subData[columns]:
        try:
            item = float(item)
            item = np.round(item,3)
            allPoints.append(item)
        except:
            next
        #allPoints.append(item)
    subData = pd.DataFrame(allPoints)
    subData.hist(bins=100)
    plt.title(columns.upper() + " Distribution")
    plt.xlabel(columns.upper())
    plt.ylabel("Frequency")
    plt.show()
```
