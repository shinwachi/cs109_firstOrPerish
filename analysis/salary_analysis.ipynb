```python
infilepath = 'filtered3_foroutliers.csv.gz'
```

```python
data = pd.read_csv(gzip.open( infilepath, 'r'), index_col=[0])
```

```python
from pandas import DataFrame
# import salary information
salary = DataFrame([dict(author=author, salary=float(salary)) for author, salary in csv.reader(open('authorSalary.txt', 'r'))])
```

```python
data = DataFrame.merge(data, salary, on='author')
```
